# Image Transformations on Lysinopril 3 photos

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import os
import PIL
from PIL import Image

import skimage.transform
import skimage.util
from skimage.exposure import adjust_gamma
```

```python
# Drugs I am using are Plavix, Effexor, Caltrate, Artorvastin, Metformin, Lysinopril

# Location of image
# I CHANGED THIS WITH EACH UNIQUE PHOTO
IMG_PATH = "images/nanas_pills/Lysinopril/Lysinopril_3.png"
```

```python
SEED = 42
random.seed(SEED)

### Load image as Numpy array

# Open image
img = PIL.Image.open(IMG_PATH)

# Convert the image to a Numpy array, optionally invert, and append to X
img = np.asarray(img)

### Get height and width of image
height = img.shape[0]
width = img.shape[1]
### View the image

# Print part of image
print(img.shape)
print("First channel:")
print(img[:,:,0])

# Display image from array
plt.imshow(img, vmin=0, vmax=255)
```

```python
### Rotate

# Create. Note that rotate() normalizes values, so we need to keep
# them in the 0..255 integer range.
rot_1 = skimage.transform.rotate(img, angle=11, mode='edge', preserve_range=True).astype(np.uint8)
rot_2 = skimage.transform.rotate(img, angle=22, mode='edge', preserve_range=True).astype(np.uint8)
rot_3 = skimage.transform.rotate(img, angle=45, mode='edge', preserve_range=True).astype(np.uint8)
rot_4 = skimage.transform.rotate(img, angle=56, mode='edge', preserve_range=True).astype(np.uint8)
rot_5 = skimage.transform.rotate(img, angle=67, mode='edge', preserve_range=True).astype(np.uint8)
rot_6 = skimage.transform.rotate(img, angle=78, mode='edge', preserve_range=True).astype(np.uint8)
rot_7 = skimage.transform.rotate(img, angle=90, mode='edge', preserve_range=True).astype(np.uint8)
rot_8 = skimage.transform.rotate(img, angle=101, mode='edge', preserve_range=True).astype(np.uint8)
rot_9 = skimage.transform.rotate(img, angle=112, mode='edge', preserve_range=True).astype(np.uint8)
rot_10 = skimage.transform.rotate(img, angle=123, mode='edge', preserve_range=True).astype(np.uint8)
rot_11 = skimage.transform.rotate(img, angle=135, mode='edge', preserve_range=True).astype(np.uint8)
rot_12 = skimage.transform.rotate(img, angle=146, mode='edge', preserve_range=True).astype(np.uint8)
rot_13 = skimage.transform.rotate(img, angle=157, mode='edge', preserve_range=True).astype(np.uint8)
rot_14 = skimage.transform.rotate(img, angle=168, mode='edge', preserve_range=True).astype(np.uint8)
rot_15 = skimage.transform.rotate(img, angle=180, mode='edge', preserve_range=True).astype(np.uint8)
rot_16 = skimage.transform.rotate(img, angle=191, mode='edge', preserve_range=True).astype(np.uint8)
rot_17 = skimage.transform.rotate(img, angle=202, mode='edge', preserve_range=True).astype(np.uint8)
rot_18 = skimage.transform.rotate(img, angle=213, mode='edge', preserve_range=True).astype(np.uint8)
rot_19 = skimage.transform.rotate(img, angle=224, mode='edge', preserve_range=True).astype(np.uint8)
rot_20 = skimage.transform.rotate(img, angle=235, mode='edge', preserve_range=True).astype(np.uint8)
rot_21 = skimage.transform.rotate(img, angle=246, mode='edge', preserve_range=True).astype(np.uint8)

# Show the original and the transforms
figs, axs = plt.subplots(nrows=1, ncols=22, figsize=(14, 14))
axs[0].imshow(img, vmin=0, vmax=255)
axs[1].imshow(rot_1, vmin=0, vmax=255)
axs[2].imshow(rot_2, vmin=0, vmax=255)
axs[3].imshow(rot_3, vmin=0, vmax=255)
axs[4].imshow(rot_4, vmin=0, vmax=255)
axs[5].imshow(rot_5, vmin=0, vmax=255)
axs[6].imshow(rot_6, vmin=0, vmax=255)
axs[7].imshow(rot_7, vmin=0, vmax=255)
axs[8].imshow(rot_8, vmin=0, vmax=255)
axs[9].imshow(rot_9, vmin=0, vmax=255)
axs[10].imshow(rot_10, vmin=0, vmax=255)
axs[11].imshow(rot_11, vmin=0, vmax=255)
axs[12].imshow(rot_12, vmin=0, vmax=255)
axs[13].imshow(rot_13, vmin=0, vmax=255)
axs[14].imshow(rot_14, vmin=0, vmax=255)
axs[15].imshow(rot_15, vmin=0, vmax=255)
axs[16].imshow(rot_16, vmin=0, vmax=255)
axs[17].imshow(rot_17, vmin=0, vmax=255)
axs[18].imshow(rot_18, vmin=0, vmax=255)
axs[19].imshow(rot_19, vmin=0, vmax=255)
axs[20].imshow(rot_20, vmin=0, vmax=255)
axs[21].imshow(rot_21, vmin=0, vmax=255)
```

```python
### Translation

# Choose random amount to translate (up to 1/4 image width, height) in either direction
tr_y = round((0.5 - random.random()) * (height / 2))
tr_x = round((0.5 - random.random()) * (width / 2))

# Perform translation to create new image
translation = skimage.transform.AffineTransform(translation=(tr_y, tr_x))
tr_1 = skimage.transform.warp(img, translation, mode='edge', preserve_range=True)
tr_1 = tr_1.astype(np.uint8)

# Choose a new random amount to translate (up to 1/4 image width, height) in either direction
tr_y = round((0.5 - random.random()) * (height / 2))
tr_x = round((0.5 - random.random()) * (width / 2))

# Perform translation to create new image
translation = skimage.transform.AffineTransform(translation=(tr_y, tr_x))
tr_2 = skimage.transform.warp(img, translation, mode='edge', preserve_range=True)
tr_2 = tr_2.astype(np.uint8)

# Show the original and the transforms
figs, axs = plt.subplots(1, 3, figsize=(10, 10))
axs[0].imshow(img, vmin=0, vmax=255)
axs[1].imshow(tr_1, vmin=0, vmax=255)
axs[2].imshow(tr_2, vmin=0, vmax=255)
```

```python
### Noise

# Add random Gaussian noise
noise_1 = skimage.util.random_noise(img, mode='gaussian', seed=SEED)
noise_1 = (noise_1 * 255).astype(np.uint8)

# Add random "salt and pepper" noise
noise_2 = skimage.util.random_noise(img, mode='s&p', seed=SEED)
noise_2 = (noise_2 * 255).astype(np.uint8)

# Show the original and the transforms
figs, axs = plt.subplots(1, 3, figsize=(10, 10))
axs[0].imshow(img, vmin=0, vmax=255)
axs[1].imshow(noise_1, vmin=0, vmax=255)
axs[2].imshow(noise_2, vmin=0, vmax=255)
```

```python
#Brightness
image_bright = adjust_gamma(img, gamma=0.25,gain=1)
image_bright2 = adjust_gamma(img, gamma=0.5,gain=1)
image_dark = adjust_gamma(img, gamma=1.5,gain=1)
image_dark2 = adjust_gamma(img, gamma=2,gain=1)
image_dark3 = adjust_gamma(img, gamma=2.5,gain=1)

fig,ax = plt.subplots(nrows=1,ncols=6,figsize=(12,12))
ax[0].imshow(img)
ax[0].set_title("Original Image", size=10)

ax[1].imshow(image_bright)
ax[1].set_title("Very Bright Image", size=10)

ax[2].imshow(image_bright2)
ax[2].set_title("Brightened Image", size=10)

ax[3].imshow(image_dark)
ax[3].set_title("Darkened Image", size=10)

ax[4].imshow(image_dark2)
ax[4].set_title("Darkened More Image", size=10)

ax[5].imshow(image_dark3)
ax[5].set_title("Darkest Image", size=10)
```
